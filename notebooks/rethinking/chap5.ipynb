```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pymc as pm
import arviz as az

# ayudantes

def estandarizar(data: np.typing.NDArray):
    """Regresa el arreglo de datos estandarizado"""
    return (data - data.mean()) / data.std(ddof=1)
```

```python
df = sns.load_dataset('penguins')

estandarizar(df['body_mass_g'])
```

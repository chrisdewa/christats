```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

k,n = 6,9
res = 200
grid = np.linspace(0, 1, res)
prior = np.ones(res)
lk = stats.binom.pmf(k=k, n=n, p=grid)
unstd_post = lk * prior
post = unstd_post / sum(unstd_post)


samples = np.random.choice(
  a=grid, 
  p=post, 
  size=int(1e4), 
  replace=True, 
)

_, axes = plt.subplots(1,2, figsize=[6, 2])
ax1, ax2 = axes.flatten()

ax1.plot(
  range(len(samples)), 
  samples, 
  'o', 
  alpha=0.1
)

ax2.plot(grid, post)
plt.tight_layout()
plt.show()
```
